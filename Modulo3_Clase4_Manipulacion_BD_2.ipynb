##**Ahora calculemos algunas tablas Pivot**

```python
table = pd.pivot_table(df, values=['humidity'], index=['country'], aggfunc="mean")
table
```

```python
table = pd.pivot_table(df, values=['humidity'], index=['country','location_name'], aggfunc="mean")
table
```

```python
table = pd.pivot_table(df, values=['humidity'], index=['country','location_name'], aggfunc="max")
table
```

```python
list_aggfunc={'median', 'mean', 'min','std'}
table = pd.pivot_table(df, values=['humidity','visibility_miles'], index=['country','location_name'], aggfunc=list_aggfunc)
table
```

```python
list_aggfunc={'median', 'mean', 'min','std'}
table = pd.pivot_table(df, values=['humidity','visibility_miles'], index=['country','location_name'], aggfunc={'humidity':['median', 'mean', 'min','std'],'visibility_miles':['median', 'mean']})
table
```
